```python
MAIN_DATA = pd.read_csv("/content/test.csv")
```

```python
df_titanic = MAIN_DATA.copy()
```

```python
df_titanic = df_titanic[["Name", "Sex",	"Age"]]
```

```python
df_titanic
```

```python
median_age = df_titanic["Age"].median()
print(f"The median age of the passengers is: {median_age}")
```

```python
df_titanic["Sex"] = df_titanic["Sex"].replace({"male": 0, "female": 1})
df_titanic
```

```python
men_count = df_titanic[df_titanic["Sex"] == 0]["Sex"].count()
women_count = df_titanic[df_titanic["Sex"] == 1]["Sex"].count()

print(f"Number of men: {men_count}")
print(f"Number of women: {women_count}")
```
